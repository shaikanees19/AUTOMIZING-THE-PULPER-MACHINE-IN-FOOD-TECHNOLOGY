```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow

# load trained model
model = YOLO("runs/detect/train/weights/best.pt")

# class names from your model
names = model.names

# map your 6 classes into 3 tomato types
class_map = {
    "Ripe": "Ripe",
    "spotted ripe": "Semi-ripe",
    "semi-ripe": "Semi-ripe",
    "unripe": "Unripe",
    "spotted unripe": "Unripe",
    "rotten": "Rotten"
}

# load image
img = cv2.imread("/content/tomato/test/images/dbe57999-Image_231.jpg") # Changed to an existing image path

# Check if image loaded successfully
if img is None:
    print("Error: Could not load image. Please check the path.")
else:
    # run detection
    results = model(img)

    for r in results:
        for box in r.boxes:
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            label = names[cls_id]

            tomato_type = class_map[label]

            print(f"Detected: {tomato_type} | confidence: {conf:.2f}")

            # draw box
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cv2.rectangle(img, (x1,y1), (x2,y2), (0,255,0), 2)
            cv2.putText(img, tomato_type, (x1, y1-10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2)

    cv2_imshow(img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
